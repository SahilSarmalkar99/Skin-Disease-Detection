```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\my notebook\nlp-aih\clinical_data.csv")
```

```python
df.head()
```

```python
x = list(df["clinical_observation"])
y = df["diagnostic_category"]
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
decode = le.inverse_transform(y)
```

```python
decode.tolist()
```

```python
y
```

```python
import nltk
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size = 500
```

```python
one_repo = [one_hot(words , voc_size) for words in x]
```

```python
one_repo
```

```python
from tensorflow.keras.layers import Embedding , Dense , Dropout , Bidirectional ,LSTM
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
set_len = 8
```

```python
emb_doc = pad_sequences(one_repo , maxlen=set_len , padding="post")
```

```python
emb_doc
```

```python
dim_size = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size , dim_size , input_length = set_len))
```

```python
model.add(Bidirectional(LSTM(100)))
model.add(Dense(5 , activation="softmax"))
```

```python
model.compile(loss="sparse_categorical_crossentropy" , optimizer="adam" , metrics=["accuracy"])
```

```python
model.summary
```

```python
x_final = np.array(emb_doc)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test ,y_train ,y_test = train_test_split(x_final , y_final , test_size=0.2 , random_state=1)
```

```python
model.fit(x_train , y_train , epochs=10 , batch_size=64)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred_classes = np.argmax(y_pred, axis=1)
```

```python
y_pred_classes
```

```python
confusion_matrix(y_test , y_pred_classes)
```

```python
accuracy_score(y_test , y_pred_classes)*100
```

```python
model.save("aih_nlp_model.h5")
```

```python
def txt_preprocessing(txt):
    one_repo = one_hot(txt , 500)
    text =  pad_sequences([one_repo] , maxlen=8 , padding="post")
    return np.array(text)
```

```python
txt = "Deep, painful cystic lesions on the jawline and back."
```

```python
result = txt_preprocessing(txt)
```
